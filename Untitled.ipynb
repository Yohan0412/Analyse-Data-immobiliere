```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Exemple de données : tu peux remplacer par ton fichier CSV ou DataFrame


df = pd.read_csv("donnees_immobilieres.csv")

# Calcul de la corrélation de Pearson
correlation = df['Surface_m2'].corr(df['Prix_euros'])
print(f"Corrélation Surface vs Prix : {correlation:.2f}")

# Graphique scatter plot avec titre indiquant la corrélation
sns.scatterplot(data=df, x='Surface_m2', y='Prix_euros')
plt.title(f"Corrélation Surface vs Prix = {correlation:.2f}")
plt.xlabel('Surface (m²)')
plt.ylabel('Prix en euro')
plt.show()
```
