# Tester la normalité des données :

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import shapiro, probplot


df = pd.read_csv("donnees_immobilieres.csv")

from scipy.stats import shapiro


# TEST Shapiro-Wilk
stat, p = shapiro(df['Surface_m2'])

print("Statistique de test :", stat)
print("p-value :", p)

if p > 0.05:
    print("✅ Les données suivent une distribution normale.")
else:
    print("❌ Les données ne suivent PAS une distribution normale.")
```

- **H₀ (hypothèse nulle)** : les données suivent une loi normale
- **H₁ (hypothèse alternative)** : les données **ne suivent pas** une loi normale
 On **rejette H₀** si la p-value est **inférieure à 0.05**.

 Étant donné que les données ne suivent pas une distribution normale, nous utiliserons le test de Spearman afin d’évaluer la corrélation.

## Calcule de la corrélation avec Spearman 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import spearmanr

# Chargement des données
df = pd.read_csv("donnees_immobilieres.csv")

# Calcul de la corrélation de Spearman
corr_spearman, p_value = spearmanr(df['Surface_m2'], df['Prix_euros'])

# Affichage des résultats
print(f"Corrélation de Spearman (Surface vs Prix) : {corr_spearman:.2f}")
print(f"Valeur p : {p_value:.4f}")

# Graphique scatter plot
sns.scatterplot(data=df, x='Surface_m2', y='Prix_euros')
plt.title(f"Corrélation de Spearman = {corr_spearman:.2f}")
plt.xlabel('Surface (m²)')
plt.ylabel('Prix en euro')
plt.show()
```

# Corrélation est de 0.79. 
Il existe donc une forte corrélation positive monotone entre la surface en m2 et le prix , donc plus le prix a tendence à augmenter aussi.

# La P-value est de 0.

Cela indique que cette corrélation est statistiquement significative. Autrement dit, il y a moins de 0,05% de chances que cette relation soit due au hasard.
